```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
# loading the dataset
df = pd.read_csv("/content/creditcard.csv")
df.head()
```

```python
df.info()
```

```python
# number of missing values
df.isnull().sum()
```

```python
df = df.dropna()
```

```python
df.shape
```

```python
# distribution of legit & fraudulent transactions
df['Class'].value_counts()
```

There is some imbalance in the data.

There are a lot more legit than fraudulent transactions and if the model is trained with this data, it wont be able to fully recognize the fraudulent ones.

0 --> Legit Transaction

1 --> Fraudulent Transaction

```python
# separating the data for analysis

legit = df[df.Class == 0]
fraud = df[df.Class == 1]
```

```python
print(legit.shape)
print(fraud.shape)
```

```python
legit.Amount.describe()
```

```python
fraud.Amount.describe()
```

```python
# compare the values of both transactions

df.groupby('Class').mean()
```

Dealing with the Unbalanced Data

Under-Sampling: Build a sample dataset containing similar distribution of legit transaction, and fraudulent transaction

Number of Fraudulent Tansactions --> 264

```python
legit_sample = legit.sample(n=264)
```

Concatenating two DataFrames

```python
new_df = pd.concat([legit_sample, fraud], axis = 0)
```

```python
new_df.head()
```

```python
new_df.tail()
```

```python
new_df['Class'].value_counts()
```

```python
new_df.groupby('Class').mean()
```

Splitting the data into Features and Targets

```python
X = new_df.iloc[:, :-1]
Y = new_df.iloc[:, -1]
print(Y)
#print(X.shape)
```

Split data inot training and test data

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, stratify=Y, random_state=2)
```

```python
X_train.shape
```

Model Training

Logistic Regression

```python
model = LogisticRegression()
```

```python
#training the logistic regression model with training data
model.fit(X_train, Y_train)
```

Model Evaluation

```python
# accuracy on training data
X_train_pred = model.predict(X_train)

training_data_accuracy = accuracy_score(X_train_pred, Y_train)

print(f"Accuracy on training data {training_data_accuracy}")
```

```python
# accuracy on test data
X_test_pred = model.predict(X_test)

test_data_accuracy = accuracy_score(X_test_pred, Y_test)

print(f"Accuracy on test data {test_data_accuracy}")
```
